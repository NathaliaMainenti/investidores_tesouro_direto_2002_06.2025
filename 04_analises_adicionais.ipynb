Gráfico Chances

```python
# ---- SEUS DADOS ----
dados = {
    "variavel": [
        "Gênero Masculino","Faixa Ano(2016-2020)","Faixa Ano[(2021-2025)",
        "Estado Civil Outro","Estado Civil Separado Divorciado","Estado Civil Solteiro",
        "Profissão Agropecuaria e Meio Ambiente","Profissão Comunicação, Artes e Midia","Profissão Direito, Segurança e Serviços Públicos",
        "Profissão Educação, Pesquisa e Ciências","Profissão Indústria, Construção e Transporte","Profissão Outros",
        "Profissão Saúde","Profissão Serviços Gerais e Atendimento","Profissão Tecnologia, Engenharia e Produção",
        "Faixa Etária 35 a 44 anos","Faixa Etária Até 24 anos","Faixa Etária Mais de 44 anos",
        "Região Nordeste","Região Norte","Região Sudeste","Região Sul"
    ],
    "valor": [
        1.286615,2.405892,5.783577,1.387897,0.904203,1.049124,1.033717,0.833796,
        0.532402,0.636070,1.107546,0.776366,0.558994,1.139511,0.526836,0.791993,
        0.823455,0.828383,1.010080,1.234624,0.786035,0.840684
    ]
}

df = pd.DataFrame(dados)

# ordenar por valor
df = df.sort_values(by="valor", ascending=False)

# definir cores por condição
colors = []
for v in df['valor']:
    if v > 1.10:
        colors.append("#154375")   # azul — aumenta odds
    elif v < 0.82:
        colors.append("#154375")   # cinza — neutro
    else:
        colors.append("#747f8e")   # vermelho — diminui odds

plt.figure(figsize=(14, 10))
bars = plt.barh(df['variavel'], df['valor'], color=colors)

# linha de referência onde OR = 1
plt.axvline(x=1, linestyle='--', color="black")

plt.gca().invert_yaxis()
plt.xlabel("Odds Ratio")

# colocar valores dentro da barra
for i, v in enumerate(df['valor']):
    pct = (v - 1) * 100
    txt = f'{v:.2f} ({pct:+.0f}%)'
    plt.text(v/2, i, txt, va='center', ha='center', fontsize=9, color="white", fontweight='bold')

plt.tight_layout()
plt.show()
```
